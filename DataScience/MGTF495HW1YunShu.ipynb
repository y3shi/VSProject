### (a) How many mutual funds have Google in its top 10 holdings? Show a fund-category-wise breakup of this total value.

```python
sql = text('''SELECT fund_category, count(*)
FROM public.mutualfunds
where top10_holdings like '%GOOGL%' and  fund_category is not null
group by fund_category;''')
result = conn.execute(sql)
for row in result:
    print(row)
```

### (b) Which growth funds has the top 3 PE Ratio values?  In a separate query, find the historical fund-return to category-return ratio for the last 5 years for these funds.

```python
sql = text('''SELECT fund_symbol,fund_price_earning_ratio
FROM public.mutualfunds
where fund_price_earning_ratio is not null
and investment_type ilike '%growth%'
order by fund_price_earning_ratio DESC
limit 3;
''')
result = conn.execute(sql)


df = pd.DataFrame(result.fetchall(),columns = result.keys())
df

```

(c) Find mutual funds not owned by the "BlackRock" family, whose fund manager went to Stanford University, and the fund has no "foreign" or "world"-level holdings.

```python
sql = text('''SELECT fund_symbol, fund_family, fund_short_name, fund_category
              FROM public.mutualfunds
              WHERE (management_bio ilike '%Stanford%') and
                    not (fund_family like '%BlackRock%')
                     and not (mutualfunds.fund_category  ilike '%foreign%'
                          or mutualfunds.fund_category ilike  '%world%');''')
result = conn.execute(sql)
df = pd.DataFrame(result.fetchall(),columns = result.keys())
df
```

(d) For funds whose 52-week high is more than $100, find the ratio between the 52-week high-low difference and the average of 52-week high and low in descending order of the ratio.

```python
sql = text('''SELECT fund_symbol,week52_high_low_change/(0.5*(week52_high+week52_low)) as ratio
from mutualfunds
where week52_high > 100
and fund_symbol is not null
order by ratio desc;''')
result = conn.execute(sql)
df = pd.DataFrame(result.fetchall(),columns = result.keys())
df
```
